```python
%%writefile app.py

import streamlit as st
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

data = pd.read_csv("/content/diabetes.csv")


x = data.loc[:,[
    'Pregnancies',
    'Glucose',
    'BloodPressure',
    'SkinThickness',
    'Insulin',
    'BMI',
    'DiabetesPedigreeFunction',
    'Age'
]]

y = data['Outcome']


x_train,x_test,y_train,y_test = train_test_split(
    x,
    y,
    test_size=0.3,
    random_state=42
)



sc = StandardScaler()

sc.fit(x_train)

x_train_std = sc.transform(x_train)

x_test_std = sc.transform(x_test)



model = Perceptron(
    max_iter=1000,
    eta0=0.05,
    random_state=42
)



model.fit(x_train_std,y_train)



y_pred = model.predict(x_test_std)

acc = accuracy_score(y_test,y_pred)



st.set_page_config(
    page_title="Diabetes Prediction",
    page_icon="🩺",
    layout="wide"
)



st.markdown(
    '''
    <style>

    .main {
        background-color: #f5f7fa;
    }

    .stButton>button {
        width: 100%;
        background-color: #ff4b4b;
        color: white;
        height: 55px;
        border-radius: 12px;
        font-size: 20px;
        border: none;
    }

    .stSlider {
        padding-top: 10px;
        padding-bottom: 10px;
    }

    </style>
    ''',
    unsafe_allow_html=True
)



st.title("🩺 HỆ THỐNG CHẨN ĐOÁN TIỂU ĐƯỜNG")

st.image(
    "https://images.unsplash.com/photo-1576091160399-112ba8d25d1d",
    use_container_width=True
)

st.markdown("---")

st.subheader("📊 Nhập thông số sức khỏe")


col1, col2 = st.columns(2)



with col1:

    preg = st.slider(
        "Pregnancies",
        0,
        20,
        1
    )

    glucose = st.slider(
        "Glucose",
        0,
        300,
        120
    )

    bp = st.slider(
        "Blood Pressure",
        0,
        200,
        70
    )

    skin = st.slider(
        "Skin Thickness",
        0,
        100,
        20
    )



with col2:

    insulin = st.slider(
        "Insulin",
        0,
        900,
        79
    )

    bmi = st.slider(
        "BMI",
        0.0,
        70.0,
        25.0
    )

    dpf = st.slider(
        "Diabetes Pedigree Function",
        0.0,
        3.0,
        0.5
    )

    age = st.slider(
        "Age",
        1,
        120,
        25
    )


if st.button("🔍 DỰ ĐOÁN NGUY CƠ"):

    input_data = [[
        preg,
        glucose,
        bp,
        skin,
        insulin,
        bmi,
        dpf,
        age
    ]]

    input_data_std = sc.transform(input_data)

    prediction = model.predict(input_data_std)

    st.markdown("---")

    if prediction[0] == 1:

        st.error("⚠️ NGƯỜI NÀY CÓ NGUY CƠ MẮC TIỂU ĐƯỜNG")

    else:

        st.success("✅ NGƯỜI NÀY KHÔNG CÓ NGUY CƠ TIỂU ĐƯỜNG")



st.markdown("---")

st.subheader("📈 ĐỘ CHÍNH XÁC MÔ HÌNH")

st.success(f"Accuracy : {acc:.2f}")


st.markdown("---")

st.subheader("📋 Các thông số được sử dụng")

st.write("""
- Pregnancies
- Glucose
- BloodPressure
- SkinThickness
- Insulin
- BMI
- DiabetesPedigreeFunction
- Age
""")
```
